```python
import os
import pandas as pd

# Definir la ruta al directorio fx_data
directorio_datos_fx = 'fx_data'

# Definir la ruta al directorio de datos procesados
directorio_datos_procesados = 'fx_data_procesados'

# Crear el directorio de datos procesados si no existe
if not os.path.exists(directorio_datos_procesados):
    os.makedirs(directorio_datos_procesados)

# Obtener la lista de archivos en el directorio fx_data
archivos = os.listdir(directorio_datos_fx)

# Iterar a través de todos los archivos en el directorio fx_data
for nombre_archivo in archivos:
    ruta_archivo = os.path.join(directorio_datos_fx, nombre_archivo)
    
    # Leer el archivo CSV en un DataFrame, usando ';' como separador
    try:
        df = pd.read_csv(ruta_archivo, sep=';')
    except Exception as e:
        print(f"Error al leer el archivo {nombre_archivo}: {e}")
        continue
    
    # Seleccionar solo las columnas 'Date', 'Close' y 'Volume'
    try:
        df = df[['Time', 'Close', 'Volume']]
    except KeyError as e:
        print(f"El archivo {nombre_archivo} no contiene las columnas 'Date', 'Close' o 'Volume': {e}")
        continue
    
    # Definir la ruta al archivo procesado
    ruta_archivo_procesado = os.path.join(directorio_datos_procesados, nombre_archivo)
    
    # Guardar el DataFrame procesado en un nuevo archivo CSV
    try:
        df.to_csv(ruta_archivo_procesado, sep=';', index=False)
        print(f"Archivo procesado guardado en: {ruta_archivo_procesado}")
    except Exception as e:
        print(f"Error al guardar el archivo procesado {nombre_archivo}: {e}")
        continue
```

```python
import pandas as pd
import os

# Directorio donde se encuentran los archivos CSV procesados
directorio = "fx_data_procesados"

# Iterar sobre todos los archivos en el directorio
for filename in os.listdir(directorio):
    if filename.endswith(".csv"):
        filepath = os.path.join(directorio, filename)

        try:
            # Leer el archivo CSV
            df = pd.read_csv(filepath)

            # Convertir la columna 'time' a formato datetime
            # Intentamos varios formatos comunes, en caso de error, se captura la excepción
            try:
                df['time'] = pd.to_datetime(df['time'], format='%Y-%m-%d %H:%M:%S') # Formato común
            except ValueError:
                try:
                    df['time'] = pd.to_datetime(df['time'], format='%Y-%m-%d %H:%M:%S.%f') # Formato con microsegundos
                except ValueError:
                    try:
                        df['time'] = pd.to_datetime(df['time'], format='%d.%m.%Y %H:%M:%S') # Formato dd.mm.yyyy
                    except ValueError:
                        try:
                            df['time'] = pd.to_datetime(df['time'], format='%d/%m/%Y %H:%M:%S') # Formato dd/mm/yyyy
                        except ValueError:
                            try:
                                df['time'] = pd.to_datetime(df['time'], format='%Y-%m-%dT%H:%M:%S.%fZ') # Formato ISO 8601 con Z
                            except ValueError:
                                try:
                                    df['time'] = pd.to_datetime(df['time'], format='%Y-%m-%dT%H:%M:%S%z') # Formato ISO 8601 con timezone offset
                                except ValueError:
                                    print(f"Error: No se pudo convertir la columna 'time' en el archivo {filename}.  Por favor, revise el formato de la fecha/hora en el archivo.")
                                    continue # Salta al siguiente archivo si no se puede convertir

            # Guardar el DataFrame de nuevo en el mismo archivo, reescribiéndolo
            df.to_csv(filepath, index=False)

            print(f"Archivo {filename} procesado y guardado correctamente.")

        except Exception as e:
            print(f"Error al procesar el archivo {filename}: {e}")

print("Proceso completado.")
```

```python
import pandas as pd
import os

# Directorio de entrada (archivos CSV procesados)
directorio_entrada = "fx_data_procesados"

# Directorio de salida (archivos CSV con 'Time' como índice)
directorio_salida = "fx_data_procesados_ti"

# Crear el directorio de salida si no existe
if not os.path.exists(directorio_salida):
    os.makedirs(directorio_salida)

# Iterar sobre los archivos en el directorio de entrada
for filename in os.listdir(directorio_entrada):
   
        filepath_entrada = os.path.join(directorio_entrada, filename)
        filepath_salida = os.path.join(directorio_salida, filename)  # Misma estructura de nombres

        try:
            # Leer el archivo CSV
            df = pd.read_csv(filepath_entrada,sep=';')

            # Intentar convertir la columna 'Time' a datetime y establecerla como índice
            try:
                df['Time'] = pd.to_datetime(df['Time'])
                df = df.set_index('Time')
            except KeyError:
                print(f"Error: La columna 'Time' no se encuentra en el archivo {filename}")
                continue  # Saltar al siguiente archivo
            except ValueError:
                print(f"Error: No se pudo convertir la columna 'Time' a datetime en el archivo {filename}.  Asegúrese de que el formato de fecha sea compatible.")
                continue # Saltar al siguiente archivo

            # Guardar el DataFrame con 'Time' como índice en el directorio de salida
            df.to_csv(filepath_salida)

            print(f"Archivo {filename} procesado y guardado en {filepath_salida}")

        except Exception as e:
            print(f"Error al procesar el archivo {filename}: {e}")

print("Proceso completado.")
```

```python
import pandas as pd
import os

# Directorio de entrada (archivos CSV con 'Time' como índice)
directorio_entrada = "fx_data_procesados_ti"

# Iterar sobre los archivos en el directorio de entrada
for filename in os.listdir(directorio_entrada):
   
        filepath_entrada = os.path.join(directorio_entrada, filename)

        try:
            # Leer el archivo CSV
            df = pd.read_csv(filepath_entrada, sep=',', index_col='Time', parse_dates=['Time'])

            # 1. Remover duplicados
            num_duplicados_antes = df.duplicated().sum()
            df.drop_duplicates(inplace=True)
            num_duplicados_despues = df.duplicated().sum()

            # 2. Manejar valores faltantes
            num_faltantes_antes = df.isnull().sum().sum()
            df.fillna(0, inplace=True)
            num_faltantes_despues = df.isnull().sum().sum()

            # Reportar los resultados
            print(f"Archivo: {filename}")
            print(f"Duplicados removidos: {num_duplicados_antes - num_duplicados_despues}")
            print(f"Valores faltantes manejados: {num_faltantes_antes - num_faltantes_despues}")
            print(f"Total de filas después de la limpieza: {len(df)}")
            print(f"Total de columnas: {len(df.columns)}")
            print("-" * 40)

        except Exception as e:
            print(f"Error al procesar el archivo {filename}: {e}")
```
